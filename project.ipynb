```python
#importing dependencies

import pandas as pd

#mysql toolkit
import pymysql #this will work as adapter
from sqlalchemy import create_engine
```

```python
print(pd.__version__)
```

```python
df = pd.read_csv('Walmart.csv', encoding_errors='ignore')

df.shape
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
#all duplicates
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
#droppping all rows with missing records
df.dropna(inplace=True)

# verify
df.isnull().sum()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df['unit_price'] = df['unit_price'].str.replace('$', '').astype(float)

df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
df['total'] = df['unit_price'] * df['quantity']
df.head()
```

**Fixing the column name to lower case**

```python
df.columns
```

```python
df.columns = df.columns.str.lower()
df.columns
```

```python
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%y')
df['Order Month'] = df['date'].dt.month 
df['Order Year'] = df['date'].dt.year
df['Order Day of Week'] = df['date'].dt.dayofweek
```

```python
df.head()
```

###  Importing Libraries

```python
import plotly.express as px #plotly.express (px): Data visualization library jo easy aur quick plots banata hai.
import plotly.graph_objects as go #plotly.graph_objects (go): Advanced and customizable graphs banane ke liye
import plotly.io as pio   #plotly.io (pio): Graph templates ko customize karne ke liye
import plotly.colors as colors 
pio.templates.default = "plotly_white"  #pio.templates.default = "plotly_white": Default theme white rakha gaya hai graphs ke liye
```

### Sales Analysis by Payment Method

```python
sales_by_category = df.groupby('payment_method')['total'].sum().reset_index()


fig = px.pie(sales_by_category, 
             values='total', 
             names='payment_method', 
             hole=0.5, 
             color_discrete_sequence=px.colors.qualitative.Pastel)

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text='Sales Analysis by Payment Method', title_font=dict(size=24))

fig.show()
```

### Monthly Sales Analysis

```python
sales_by_month = df.groupby('Order Month')['total'].sum().reset_index()
fig = px.line(sales_by_month, 
              x='Order Month', 
              y='total', 
              title='Monthly Sales Analysis')
fig.show()
```

### Sales Analysis by Category

```python
sales_by_category = df.groupby('category')['total'].sum().reset_index()


fig = px.pie(sales_by_category, 
             values='total', 
             names='category', 
             hole=0.5, 
             color_discrete_sequence=px.colors.qualitative.Pastel)

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text='Sales Analysis by Category', title_font=dict(size=24))

fig.show()
```

### Sales Analysis by City

```python
sales_by_subcategory = df.groupby('city')['total'].sum().reset_index()
fig = px.bar(sales_by_subcategory, 
             x='city', 
             y='total', 
             title='Sales Analysis by City')
fig.show()
```

#### Profit column creation

```python
df['Profit'] = df['profit_margin']*df['total']
```

### Monthly Profit Analysis

```python
profit_by_month = df.groupby('Order Month')['Profit'].sum().reset_index()
fig = px.line(profit_by_month, 
              x='Order Month', 
              y='Profit', 
              title='Monthly Profit Analysis')
fig.show()
```

### Profit Analysis by Category

```python
profit_by_category = df.groupby('category')['Profit'].sum().reset_index()

fig = px.pie(profit_by_category, 
             values='Profit', 
             names='category', 
             hole=0.5, 
             color_discrete_sequence=px.colors.qualitative.Pastel)

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text='Profit Analysis by Category', title_font=dict(size=24))

fig.show()
```

### Profit Analysis by City

```python
profit_by_subcategory = df.groupby('city')['Profit'].sum().reset_index()
fig = px.bar(profit_by_subcategory, x='city', 
             y='Profit', 
             title='Profit Analysis by City')
fig.show()
```
